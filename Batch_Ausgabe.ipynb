```python
import matplotlib.pyplot as plt
x = [58, 53, 48, 45]
y = [0.93792, 0.93811, 0.93796, 0.93765]
sizes = [64, 128, 256, 512] # Size per point
#plt.ylim(0,max(y))  
#plt.xlim(0, 1) 

plt.plot(x,y, linestyle='-', marker='',color='black')
plt.scatter(x, y, s=sizes, linewidths=1, color='red', label = "batch-size [512, 256, 128, 64]")
plt.title("Batch Sizes")
plt.ylabel('Accuracy')
plt.xlabel('Minutes per Epoch')
plt.legend(loc="lower right")
plt.show()
#legend 
```
